funcion fibonacci mas sencilla

```python
def fibonacci(hasta):
    primero = 0
    segundo = 1
    for i in range(hasta):
        primero, segundo = segundo, primero + segundo
    return primero
```

funcion para calcular tiempo promediando cada iteracion 20 veces

```python
def tiempo(hasta):
    promedio=0
    for i in range(20):
        t0 = time.clock()
        fibonacci(hasta)
        t1 = time.clock()
        promedio += (t1-t0)
        #print ("timepo de cada i",t1-t0)
    return promedio/20

```

grafica de la funcion fibonacci hasta la iteracion 5000

Grafica

```python
hasta=5000
X = np.arange(hasta)
Y = [tiempo(i) for i in range(hasta)]
plt.xlabel('Hasta que numero')
plt.ylabel('Tiempo(ms)')
plt.plot(X,Y)
```
